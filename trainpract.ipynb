```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('train.csv')
data.head(10)
```

```python
data.info() #getting information about the data
```

```python
data.dropna(how = "all",inplace = True ) #to drop the rows have null values instead of all any can also be used
data.head(100)
```

```python
X  = data.iloc[:,0:10].values #storing the input values
Y= data.iloc[:,10].values #storing output values
```

```python
labelencoder_X = LabelEncoder()
X[:,3] = labelencoder_X.fit_transform(X[:,3])
ct = ColumnTransformer([("Course_Domain", OneHotEncoder(), [3])],    remainder = 'passthrough')
X = ct.fit_transform(X)
```
